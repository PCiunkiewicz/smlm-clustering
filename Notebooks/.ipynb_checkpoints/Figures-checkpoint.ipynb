```python
df = pd.read_csv('Training_data/Training_EASY.csv', header=0, index_col=0)
XY = df.values

hdb = hdbscan.HDBSCAN(core_dist_n_jobs=6, min_cluster_size=400,
                      min_samples=50, gen_min_span_tree=True)
_ = hdb.fit(XY)

nclusters = len(set(hdb.labels_[hdb.labels_ != -1]))
setattr(hdb, 'n_clusters_', nclusters)

sil_samples = silhouette_samples(XY, hdb.labels_)
setattr(hdb, 'silhouette_samples_', sil_samples)
setattr(hdb, 'silhouette_score_', sil_samples.mean())
setattr(hdb, 'n_samples_', XY.shape[0])
```

```python
sns.set()
from yellowbrick.cluster import SilhouetteVisualizer
visualizer = SilhouetteVisualizer(hdb)

visualizer.draw(hdb.labels_)
# visualizer.finalize()
plt.ylabel('Cluster label')
plt.xlabel('Silhouette coefficient')
plt.yticks([])
plt.savefig('test.png', bbox_to_inches='tight', rasterized=True, dpi=100)
```
